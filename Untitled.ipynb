# Problem Statement
Resale and marketplace platforms show large price variations for similar products, making it difficult for buyers and resellers to judge whether a listed price is fair. 
There is no automated system that evaluates pricing fairness using market-driven patterns. 
This project builds a machine learningâ€“based pricing intelligence system that learns from real marketplace data to estimate a fair market price. 
By comparing the predicted fair price with the listed price, the system classifies items as underpriced, fairly priced, or overpriced. 
The goal is to support better buying, reselling, and deal-detection decisions using data-driven insights.


# Metrics Evaluation
Root Mean Squared Error (RMSE) is chosen as the evaluation metric because it directly measures the average magnitude of prediction errors in the same units as price. 
RMSE penalizes larger errors more heavily, which is important for pricing models where big mistakes are costly. 
It provides a clear and interpretable measure of how close predicted prices are to actual prices. 
RMSE is widely used for regression problems involving continuous numerical targets. 
This makes it well-suited for evaluating pricing fairness models.


# API calling


```python
import requests
import pandas as pd

all_products = []

for skip in range(0, 1000, 100):
    url = f"https://dummyjson.com/products?limit=100&skip={skip}"
    data = requests.get(url).json()
    all_products.extend(data["products"])

rows = []

for item in all_products:
    rows.append({
        "title": item.get("title"),
        "brand": item.get("brand"),
        "category": item.get("category"),
        "price": item.get("price"),
        "rating": item.get("rating"),
        "stock": item.get("stock")
    })

df = pd.DataFrame(rows)
df.to_csv(r"C:\Users\HP\OneDrive\Desktop\ml hackthon project\marketplace_data.csv", index=False)

print(df.shape)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

path = r"C:\Users\HP\OneDrive\Desktop\ml hackthon project\marketplace_data.csv"
base_path = r"C:\Users\HP\OneDrive\Desktop\ml hackthon project\\"

df = pd.read_csv(path)

df['price_bin'] = pd.qcut(df['price'], q=5, labels=False)

X = df.drop(['price', 'price_bin'], axis=1)
y = df[['price']]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=df['price_bin'],
    random_state=42
)

X_train.to_csv(base_path + "X_train.csv", index=False)
X_test.to_csv(base_path + "X_test.csv", index=False)
y_train.to_csv(base_path + "y_train.csv", index=False)
y_test.to_csv(base_path + "y_test.csv", index=False)

df.to_csv(base_path + "full_dataset_with_bins.csv", index=False)
```

To ensure a fair and unbiased model evaluation, the dataset is split using stratified sampling based on binned price values. 
This approach preserves the distribution of low, medium, and high-priced items in both training and testing sets. 
The data is then separated into independent feature and target files for clean experimentation. 
Saving the splits as separate CSV files improves reproducibility and modularity of the machine learning pipeline. 
This structure allows independent preprocessing, training, and evaluation without data leakage.

Exploratory analysis shows that pricing behavior varies significantly across categories and is not linearly dependent on individual features such as rating or stock. This validates the need for a machine learningâ€“based pricing fairness model that learns complex, category-aware patterns from real marketplace data.

```python
import pandas as pd
import numpy as np

path = r"C:\Users\HP\OneDrive\Desktop\ml hackthon project\marketplace_data.csv"
df = pd.read_csv(path)

category_avg = df.groupby('category')['price'].mean()
category_median = df.groupby('category')['price'].median()
category_count = df['category'].value_counts()
brand_count = df['brand'].value_counts()

df['category_avg_price'] = df['category'].map(category_avg)
df['category_median_price'] = df['category'].map(category_median)
df['price_deviation_from_category'] = df['price'] - df['category_avg_price']
df['price_percentile_in_category'] = df.groupby('category')['price'].rank(pct=True)

df['brand_trend_score'] = df['brand'].map(brand_count).fillna(1)
df['category_trend_score'] = df['category'].map(category_count)

df['demand_score'] = df['category_trend_score'] * df['rating']
df['estimated_time_to_sell'] = 30 / (df['demand_score'] + 1)

df['predicted_market_price'] = df['category_avg_price']
df['pricing_gap'] = df['predicted_market_price'] - df['price']

df['platform_fee_estimate'] = df['predicted_market_price'] * 0.08
df['expected_resale_profit'] = df['predicted_market_price'] - df['price'] - df['platform_fee_estimate']

df.to_csv(r"C:\Users\HP\OneDrive\Desktop\ml hackthon project\feature_engineered_data.csv", index=False)

df.head()
```
